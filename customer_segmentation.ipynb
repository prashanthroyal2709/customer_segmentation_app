```python
Final_df=pd.get_dummies(Final_df,columns=['Education','Marital_Status'],drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numeric_cols = ['Income','Recency','NumWebVisitsMonth','Complain','Response',
                'Age','Children','Products','Total_purchases','Acceptedcmp','since']
Final_df[numeric_cols]=scaler.fit_transform(Final_df[numeric_cols])
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
data_pca=pca.fit_transform(Final_df[numeric_cols].values)
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=data_pca[:,0],y=data_pca[:,1],c=data_pca[:,2])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init="k-means++")
```

```python
labels=kmeans.fit_predict(data_pca)
```

```python
import pandas as pd
import joblib

# 🔹 Load saved models
kmeans_loaded = joblib.load('kmeans_customer_segmentation.pkl')
scaler_loaded = joblib.load('scaler.pkl')
pca_loaded = joblib.load('pca_model.pkl')
final_columns = joblib.load('final_df_columns.pkl')

# 🔹 Numeric columns to scale
numeric_cols = ['Income','Recency','NumWebVisitsMonth','Complain','Response',
                'Age','Children','Products','Total_purchases','Acceptedcmp','since']

# 🔹 Function to predict customer cluster
def predict_customer_segment(new_data):
    # 1️⃣ Add missing columns with default value 0
    for col in final_columns:
        if col not in new_data.columns:
            new_data[col] = 0
    
    # 2️⃣ Reorder columns to match training data
    new_data = new_data[final_columns]
    
    # 3️⃣ Scale numeric columns
    new_data[numeric_cols] = scaler_loaded.transform(new_data[numeric_cols])
    
    # 4️⃣ Apply PCA on numeric columns
    new_data_pca = pca_loaded.transform(new_data[numeric_cols].values)
    # 5️⃣ Predict cluster
    cluster_label = kmeans_loaded.predict(new_data_pca)
    
    return cluster_label

# 🔹 Example new customer
new_customer = pd.DataFrame({
    'Income':[50000], 'Recency':[10], 'NumWebVisitsMonth':[5],
    'Complain':[0], 'Response':[1], 'Age':[35], 'Children':[2],
    'Products':[3], 'Total_purchases':[2000], 'Acceptedcmp':[1], 'since':[24]
})

# 🔹 Predict cluster
print("Predicted cluster:", predict_customer_segment(new_customer)[0])
```
